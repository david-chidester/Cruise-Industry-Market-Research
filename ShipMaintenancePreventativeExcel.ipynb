# Let's Look at Call type cost avgs by month and day

```python
monthCallPrice = dfShip.groupby(['Fiscal Period', 'Call Type'])['priceTotal'].mean()
monthCallPrice.head(20)
```

And **here** is a count of the *unique* repairs per fiscal period

```python
dfShip["Fiscal Period"].value_counts()
```

## Now compare the value counts to the unique graphs of cost averages per month

```python
fig, ax = plt.subplots(figsize=(15, 7))
plt.title("Average Ship Maintenance Cost by Type per Month", fontsize=21)
plt.ylabel("Average Maintenance Cost", fontdict=None, labelpad=None, fontsize=16)
plt.xlabel('month', fontsize=16)
# fig.set_size_inches(11.7, 8.27)
monthCallPrice.unstack().plot(ax=ax)
plt.xticks(rotation=90)
plt.show()
```
